```python
from pydantic import BaseModel, Field
class User(BaseModel):
    id: int
    name: str
    
u = User(id="123", name="Alice") # 自動把字串 "123" 轉成 int
print(u.id, type(u.id))
```

可以設定預設值 (Default Values)

```python
class User(BaseModel):
    name: str
    age: int = 18
    
u = User(name="Alice")
print(u.age)
```

可以嵌套模型 (Nested Models)

```python
class Metadata(BaseModel):
    source: str
    type: str
    
class Result(BaseModel):
    page_content:str
    metadata: Metadata
    
r = Result(page_content="Some content",
           metadata={"source":"file.md", "type":"emplyees"})

print(r.metadata.source) # file.md
```

```python
r
```

自動轉成字典 / JSON

```python
print(r.model_dump()) # 轉成 dict
print(r.model_dump_json()) # 轉成 JSON 字串
```

支援 alias（別名），有時候資料來源的欄位名稱不一樣，可以用 alias：

```python
class User(BaseModel):
    user_name: str = Field(alias="username")
    
data = {"username": "Alice"}
u = User(**data)  # **data 等於 "username"="Alice"
print(u.user_name)
```
